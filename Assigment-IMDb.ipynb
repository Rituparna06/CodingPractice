```python
# Create the dataframe df_by_genre

df_by_g = df2.iloc[:,11:60]
```

```python
# Add the grouped data frames and store it in a new data frame


df_a = df_by_g_1.add(df_by_g_2,fill_value=0)
df_a = df_a.add(df_by_g_3,fill_value=0)
```

```python
# Extract genres with atleast 10 occurences


g_top10 = df_a[df_a.c > 10]
g_top10
```

```python
# Take the mean for every column by dividing with cnt 

g_top10[g_top10.columns[0:-1]] = g_top10[g_top10.columns[0:-1]].div(g_top10.c, axis=0)
```
